# Main code block to direct the Robomaster to search for the doll wearing clothes that match the description.

```python
# global variables
arenadimensions = (3.86, 2.63)
mazedimensions = (2.40, 1.60)

startingpos = [0.5*arenadimensions[0],0,0]

# search coordinates
arenacenter = (0.5*arenadimensions[0], 0.5*arenadimensions[1])

# leave 20cm space between doll and robot
searchboundary = {"left": {"start":(0.1*arenadimensions[0], 0.2*arenadimensions[1]), "end":(0.1*arenadimensions[0], 0.95*arenadimensions[1]), "searched":False}, 
                "center": {"start":(0.2*arenadimensions[0], 0.95*arenadimensions[1]), "end":(0.8*arenadimensions[0], 0.95*arenadimensions[1]), "searched":False},
                "right":{"start":(0.90*arenadimensions[0], 0.8*arenadimensions[1]), "end":(0.90*arenadimensions[0], 0.3*arenadimensions[1]), "searched":False}}

# victim description
victim = ["dresses"]#["tops", "trousers"]

# define functions

def movexz(currentpos, x, z=None): # param = {"x": xx, "y": yy, z": zz}    
    # take angles around the robot on command
    
#     print("movexz: original position", currentpos) 
        # reset z axis
#     print("movexz: before reset z", currentpos)
    if currentpos[2] > 180 and currentpos[2] < 360:
        currentpos[2] =  - (180 - currentpos[2] % 180)

    elif currentpos[2] < -180 and currentpos[2] > -360:
        currentpos[2] = currentpos[2] % 180 

    elif currentpos[2] >= 360:
        currentpos[2] = currentpos[2] % 360
    
    elif currentpos[2] <= -360:
        currentpos[2] = -(currentpos[2] % 360)

#     print("movexz: after reset z", currentpos)

    # rotate
    if z != None:
        robot._sendcommand("chassis move z " + str(z)) # + " vz 10")
        currentpos[2] += z
        time.sleep(abs(z / 20))
        
    # reset z axis
#     print("movexz: before reset z", currentpos)
    if currentpos[2] > 180 and currentpos[2] < 360:
        currentpos[2] =  - (180 - currentpos[2] % 180)

    elif currentpos[2] < -180 and currentpos[2] > -360:
        currentpos[2] = currentpos[2] % 180 

    elif currentpos[2] >= 360:
        currentpos[2] = currentpos[2] % 360
    
    elif currentpos[2] <= -360:
        currentpos[2] = -(currentpos[2] % 360)
    
#     print("movexz: after reset z", currentpos)
    # update x,y coordinates according move
    if currentpos[2] >= 0 and currentpos[2] < 90: #TR 
        theta = math.radians(currentpos[2])
        currentpos[0] += abs(x * math.sin(theta))
        currentpos[1] += abs(x * math.cos(theta))

    elif currentpos[2] >= 90 and currentpos[2] < 180: #BR
        theta = 180 - currentpos[2]
        theta = math.radians(theta)
        currentpos[0] += abs(x * math.sin(theta))
        currentpos[1] -= abs(x * math.cos(theta))

    elif currentpos[2] < 0 and currentpos[2] > -90: #TL
        theta = abs(currentpos[2])
        theta = math.radians(theta)
        currentpos[0] -= abs(x * math.sin(theta))
        currentpos[1] += abs(x * math.cos(theta))

    elif currentpos[2] < -90 and currentpos[2] > -180: #BL
        theta = 180 - abs(currentpos[2])
        theta = math.radians(theta)
        currentpos[0] -= abs(x * math.sin(theta))
        currentpos[1] -= abs(x * math.cos(theta))

    elif currentpos[2] == -90:
        currentpos[0] -= abs(x)

    elif currentpos[2] == 90:
        currentpos[0] += abs(x)

    elif currentpos[2] == 0:
        currentpos[1] += abs(x)

    elif currentpos[2] == abs(180):
        currentpos[1] -= abs(x)
        
    print("movexz: new positions:", currentpos)
    # move
    cmd = "x {}".format(x)
    robot._sendcommand("chassis move " + cmd) # + " vxy 0.3")
    time.sleep(abs(x / 0.4))
    
    return currentpos

def movecoord(currentpos, c_coord, pointwhere=0):
    # take position of the robot as actual coordinates on grid
#     print("movecoord:", pointwhere)
    # rotate to 0 and update coordinates
    currentpos = movexz(currentpos, 0, 0 - currentpos[2])
    
    tr = None
    br = None
    tl = None
    bl = None
    
    c_x = c_coord[0]
    c_y = c_coord[1]
    x = currentpos[0]
    y = currentpos[1]
    
    if y == c_y and x == c_x:
        return currentpos
    elif x == c_x and y < c_y:
        h = abs(y - c_y)
        currentpos = movexz(currentpos, h)

    elif x == c_x and y > c_y:
        h = abs(y - c_y)
        currentpos = movexz(currentpos, h, 180)

    elif y == c_y and x < c_x:
        h = abs(x - c_x)
        currentpos = movexz(currentpos, h, 90)

    elif y == c_y and x > c_x:
        h = abs(x - c_x)
        currentpos = movexz(currentpos, h, -90)
        
    else:
        theta = math.atan(abs((x - c_x) / (y - c_y)))
        h = abs(x - c_x) * math.sin(theta)
        theta = math.degrees(theta)
#         print(theta, h)
        
        # find quadrant
        if c_x > x and c_y < y: #TL
            theta = 180 - theta
            
        elif c_x < x and c_y < y: #TR
            theta = -(180 - theta)
            
        elif c_x > x and c_y > y: #BL
            theta = theta
            
        elif c_x < x and c_y > y: #BR
            theta = -theta

        currentpos = movexz(currentpos, h, theta)
    # point to direction indicated, default=0
    if pointwhere != None:
        currentpos = movexz(currentpos, 0, pointwhere - currentpos[2])
    print("movecoord:", currentpos)
    return currentpos

def slideLR(currentpos, mag): # slides from left to right 
    z = currentpos[2] 
    
    robot._sendcommand("chassis move y {}".format(mag))
    time.sleep(abs(mag / 0.4))

    if abs(z - 90) < 1:
        currentpos[1] -= mag
    elif abs(z - 0) < 1:
        currentpos[0] += mag
    elif abs(z + 90) < 1:
        currentpos[1] += mag
    
    print("slideLR:", currentpos)
    return currentpos

def centerDoll(currentpos, x):    
    if x < 0.5: # doll is on the left
        currentpos = slideLR(currentpos, -0.1)
    elif x > 0.5: # doll is on the right
        currentpos = slideLR(currentpos, 0.1)
        
    return currentpos

def adjDist(currentpos, h):
    if h > 0.3: # if doll is too near
        currentpos = movexz(currentpos, -0.05, 0)
    elif h < 0.2: # if doll is too far
        currentpos = movexz(currentpos, 0.05, 0)
    return currentpos

def preparePickup(currentpos, h):
    if h < 0.4: # if doll not in claw range
        currentpos = movexz(currentpos, 0.02, 0)
    elif h > 0.46: # claw overshot doll
        currentpos = movexz(currentpos, -0.02, 0)
    return currentpos

class VideoScreenshot(object):
    def __init__(self, src=0):
        # Create a VideoCapture object
        self.capture = cv2.VideoCapture(src)

        # Take screenshot every x seconds
        self.screenshot_interval = 1

        # Default resolutions of the frame are obtained (system dependent)
        self.frame_width = int(self.capture.get(3))
        self.frame_height = int(self.capture.get(4))

        # Start the thread to read frames from the video stream
        self.thread = Thread(target=self.update, args=())
        self.thread.daemon = True
        self.thread.start()

    def update(self):
        # Read the next frame from the stream in a different thread
        while True:
            if self.capture.isOpened():
                (self.status, self.frame) = self.capture.read()

    def show_frame(self):
        # Display frames in main program
        if self.status:
            cv2.imshow('frame', self.frame)

        # Press Q on keyboard to stop recording
        key = cv2.waitKey(1)
        if key == ord('q'):
            self.capture.release()
            cv2.destroyAllWindows()
            exit(1)

    def save_frame(self):
        # Save obtained frame periodically
        self.frame_count = 0
        def save_frame_thread():
            while True:
                try:
                    name = './ss/' + str(self.frame_count) + '.jpg'
                    cv2.imwrite(name, self.frame)
                    self.frame_count += 1
                    time.sleep(self.screenshot_interval)
                except AttributeError:
                    pass
        Thread(target=save_frame_thread, args=()).start()

if __name__ == '__main__':
    # initialise YOLOv4 model
    print("\nLoading YOLOv4 model configuration...")
    yolo = tf.YOLOv4()
    yolo.classes = "yolov4_cfg/obj.names"
    classes = yolo.classes # define all classes
    yolo.make_model() # create model
    yolo.load_weights("yolov4_cfg/til5k.weights", weights_type="yolo") # load weights

    # initialise robot
    print("Initialising robot...")
    robot = Robot(findrobotIP())
    robot._sendcommand('led control comp bottom_all r 255 g 255 b 255 effect solid')
    robot._sendcommand('stream on')

    # initialise video capture
    print("Initialising video capture...")
    rtsp_stream_link = f"tcp://@192.168.1.6:40921"
    video_stream_widget = VideoScreenshot(rtsp_stream_link)
    video_stream_widget.save_frame()
    
    # initialise objectives
    pickedup = False
    # ---------------------------------------------------------------------------------
    # main algorithm
    
    pos = startingpos
    # move to center 
    pos = movecoord(startingpos, arenacenter)

    # start parameter search
    found = False
    
    # define search areas
    searcharea = ("left", "center", "right")
    direction = None
    mag = None

    for area in searcharea:
        # print(area)
        robot._sendcommand('led control comp bottom_all r 255 g 255 b 255 effect solid')
        start = searchboundary[area]["start"]
        print(start)
        end = searchboundary[area]["end"]
        print(end)
        if area == "left":
            direction = -90
            mag = 0.1*arenadimensions[1]
        elif area == "center":
            direction = 0
            mag = 0.1*arenadimensions[0]
        elif area == "right":
            direction = 90
            mag = 0.1*arenadimensions[1]
            
        # move to search area start boundary
        if (abs(pos[0] - start[0]) < 0.1 and abs(pos[1] - start[1]) < 0.1) and abs(pos[2] - direction) > 1:
            pos = movexz(pos, 0, direction - pos[0])
            
        elif abs(pos[0] - start[0]) > 0.1 and abs(pos[1] - start[1]) > 0.1:
            pos = movecoord(pos, start, direction)
        
        else:
            pos = movecoord(pos, start, direction)
        
        # do object detection
        while not found and not searchboundary[area]["searched"]:
            
            if area == "left" and pos[1] >= end[1]:
                break
            elif area == "center" and pos[0] >= end[0]:
                break
            elif area == "right" and pos[1] <= end[1]:
                break
            
            if searchboundary[area]["searched"] == True:
                break
                
            pos = slideLR(pos, mag)
            time.sleep(3)
            # get latest screenshot
            list_of_files = glob.glob('ss/*')
            latest_file = max(list_of_files, key=os.path.getctime)

            # inferencing on image
            preds, bboxes = yolo.inference(media_path=latest_file)
            print("potential predictions:", preds, bboxes)
            if bboxes[0][-1] != 0: # if positive predictions exist
                preds_conv = [classes[i] for i in preds]
                print("possible predictions:", preds_conv)

                # go nearer to get clearer picture and read another screenshot to double confirm
                savepos = pos
                x = bboxes[0][0]
                h = bboxes[0][3]

                try: # adjust x position
                    while not (x < 0.6 and x > 0.49): # center doll to get a clearer detection
                        pos = centerDoll(pos, x)
                        time.sleep(2)
                        list_of_files = glob.glob('ss/*')
                        latest_file = max(list_of_files, key=os.path.getctime)
                        preds, bboxes = yolo.inference(media_path=latest_file)
                        print("x adjust:", preds, bboxes)
                        x = bboxes[0][0]
                        h = bboxes[0][3]
                except KeyboardInterrupt:
                    pass

                try: # adjust distance from doll
        #             while not (h > 0.2 and h < 0.3): # adjust distance to get a clearer detection
                    while True:
                        pos = adjDist(pos, h)
                        time.sleep(2)
                        list_of_files = glob.glob('ss/*')
                        latest_file = max(list_of_files, key=os.path.getctime)
                        preds, bboxes = yolo.inference(media_path=latest_file)
                        print("h adjust:", preds, bboxes)
                        h = bboxes[0][3]
                except KeyboardInterrupt:
                    pass

                # doll should at optimal position for inferencing    
                list_of_files = glob.glob('ss/*')
                latest_file = max(list_of_files, key=os.path.getctime)
                preds, bboxes = yolo.inference(media_path=latest_file)

                preds_conv = [classes[i] for i in preds] 
                if set(preds_conv) == set(victim): # if victim found
                    found = True
                    robot._sendcommand('led control comp bottom_all r 0 g 255 b 0 effect scrolling')
                    print("found em")
                    break
                else:
                    robot._sendcommand('led control comp bottom_all r 255 g 0 b 0 effect scrolling')
        #             pos = movecoord(pos, savepos, savepos[2]) # if victim not found, go back to original position and resume search
                    searchboundary[area]["searched"] = True # stop the while loop and move on to the next search area
                    pos = movecoord(pos, arenacenter)
                    
            else:
                print("no predictions")
        if found == True:
            print("search stopped")
            # victim identified, proceed with pickup
            try:
                while not (h >= 0.4 and h <= 0.45):
                    pos = preparePickup(pos, h)
                    time.sleep(2)
            except KeyboardInterrupt:
                pass

            while not pickedup:
                robot.closearm()
                time.sleep(1)

                check = input("Pickedup?")

                if check == "n":
                    robot.openarm()
                    time.sleep(1)
                elif check == "y":
                    pickedup = True

            if pickedup:
                print("congrats")
                break
```
