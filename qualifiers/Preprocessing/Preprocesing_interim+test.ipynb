```python
import PIL
import pandas 
import os
from PIL import Image
import shutil
```

<h1>Split the test images up<h1>

```python
imgdir = "../test" # change this to the folder containing the test images
```

```python
imgs = os.listdir(imgdir)
```

```python
imgs[:5] # you should see the image names here
```

```python
# you should see 1.5k here
print(len(imgs))
```

```python
split = [imgs[i*329:i*329 + 329] for i in range(6)]
```

```python
len(split)
```

```python
# splitting the image folder into 5 folders and copying images into the folders
team = ["weiching", "weichen", "xiaoyue", "genice", "weihao", "sean"]

for i in team:
    dest = os.path.join(imgdir, i)
    if not os.path.exists(dest):
        os.mkdir(dest)
    idx = team.index(i)
    allocated_imgs = split[idx]
    for j in allocated_imgs:
        shutil.copy2(os.path.join(imgdir, j), dest)
    shutil.make_archive(dest, 'zip', dest)
```
