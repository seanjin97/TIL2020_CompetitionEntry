<h3>Change category IDs<h3>

```python
d = {0:2, 1:0, 2:1, 3:3, 4:4}
```

```python
texts = os.listdir("../til2020/newlabels/interim/fortrain")
len(texts)
```

```python
[i for i in texts if not i.endswith("txt")]
```

```python
for i in texts:
    with open(os.path.join("../til2020/newlabels/interim/fortrain", i), "r") as f:
        l = []
        for line in f:
            x = line.strip().split()
            x[0] = str(d[int(x[0])])
            s = " ".join(x) + "\n"
            l.append(s)
    with open(os.path.join("../til2020/newlabels/interim/fortrain", i), "w") as f:
        f.writelines(l)
```
